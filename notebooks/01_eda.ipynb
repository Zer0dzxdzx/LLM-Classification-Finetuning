Understand the competition files, labels, text lengths, and obvious data-quality issues.

```python
from pathlib import Path
import pandas as pd

DATA_DIR = Path('/kaggle/input/llm-classification-finetuning')
if not DATA_DIR.exists():
    DATA_DIR = Path('../data/sample')

train = pd.read_csv(DATA_DIR / 'train.csv')
test = pd.read_csv(DATA_DIR / 'test.csv')
train.shape, test.shape
```

```python
train.head()
```

```python
LABELS = ['winner_model_a', 'winner_model_b', 'winner_tie']
train[LABELS].sum().sort_values(ascending=False)
```

```python
text_cols = ['prompt', 'response_a', 'response_b']
lengths = train[text_cols].fillna('').astype(str).applymap(len)
lengths.describe(percentiles=[0.5, 0.75, 0.9, 0.95, 0.99])
```

```python
train[text_cols].isna().sum()
```
